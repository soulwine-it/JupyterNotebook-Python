```python
import cv2
import matplotlib.pyplot as plt

# 이미지 원본
image = cv2.imread('park.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# 이미지 확대
expand = cv2.resize(image, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(expand, cv2.COLOR_BGR2RGB))
plt.show()

# 이미지 축소
shrink = cv2.resize(image, None, fx=0.8, fy=0.8, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(shrink, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import numpy as np

image = cv2.imread('park.jpg')
# 행과 열 정보만 저장합니다.
height, width = image.shape[:2]

M = np.float32([[1, 0, 50], [0, 1, 10]]) # 50만큼 우측으로 이동  10만큼 아래로 이동
dst = cv2.warpAffine(image, M, (width, height))

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('park.jpg')

# getRotationMatrix2D(width / 2, height / 2), 90, 0.5)
#- center : 회전중심 - angle : 회전 각도 - scale : Scale Factor
# 행과 열의 정보만 저장
height, width = image.shape[:2]

M = cv2.getRotationMatrix2D((width / 2, height / 2), 45, 0.5) # 45도 각도로 하고 스케일 값을 반으로 하여 크기를 반으로 줄임
print(M)
dst = cv2.warpAffine(image, M, (width, height))

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
